```python
import requests as rq
from bs4 import BeautifulSoup
import pandas as pd
```

```python
import io

html = rq.request('GET',"https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-PY0220EN-SkillsNetwork/labs/project/netflix_data_webpage.html")

soup = BeautifulSoup(html.text, 'html.parser')

# Use a CSS selector to find the table with data-test="historical-prices"
table = soup.find_all(attrs={"data-test": "historical-prices"})

# Check if the table is found and print it
if table:
    print("Table found successfully!")
    # print(table[0].prettify()) # Print the first found table in a readable format
else:
    print("Table not found.")

# Read the HTML table into a DataFrame and assign it directly
# Wrap the HTML string in StringIO to address the FutureWarning
frame1 = pd.read_html(io.StringIO(str(table[0])), flavor = 'bs4')[0]

# Display the head of the DataFrame
display(frame1.head())
```

**Another way to do webscraping **

```python
import requests as rq
from bs4 import BeautifulSoup
import pandas as pd

html = rq.request('GET',"https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBMDeveloperSkillsNetwork-PY0220EN-SkillsNetwork/labs/project/netflix_data_webpage.html")

soup = BeautifulSoup(html.text, 'html.parser')

#table = soup.find_all(attrs={"data-test": "historical-prices"})
netflix_stock = pd.DataFrame(columns=["Date","Open","High"])


for row in soup.find("tbody").find_all('tr'):
  r = row.find_all('td')
  date = r[0].text
  open = r[1].text
  high = r[2].text

  #netflix_data = pd.concat([netflix_data,pd.DataFrame({"Date":[date], "Open":[Open], "High":[high], "Low":[low], "Close":[close], "Adj Close":[adj_close], "Volume":[volume]})], ignore_index=True)

  netflix_stock = pd.concat([netflix_stock,pd.DataFrame({"Date": [date],"Open":[open],"High":[high]})], ignore_index=True)

netflix_stock.head()
```

```python
last_open_value = netflix_stock['Open'].iloc[-1]
print(f"The last 'Open' value is: {last_open_value}")
```
